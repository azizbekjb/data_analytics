```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Chastotalar poligoni

```python
x_i = [2, 3, 5, 6]
n_i = [10, 15, 5, 20]
plt.plot(x_i, n_i, color='red')
plt.xlabel('X o\'qi')
plt.ylabel('Chastotalar')
plt.title('Chastotalar poligoni')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid(axis='x', linestyle='--', alpha=0.7)
df = pd.DataFrame({
    'x_i' : x_i,
    'n_i' : n_i
})
```

Chastotalar gistogrammasi

```python
bosh_qiymat = 3
qadam = 2
sikl = 7
arr = []
for i in range(sikl):
    arr.append(bosh_qiymat)
    bosh_qiymat += qadam

n_i = [4, 6, 20, 40, 20, 4, 6]
chst_dev_h = [i/qadam for i in n_i]

df = pd.DataFrame({
    'x_i': arr,
    'n_i': n_i,
    'n_i/h': chst_dev_h
})

plt.figure(figsize=(6,6))
plt.bar(df['x_i'], df['n_i/h'], width=qadam, align='edge', edgecolor='red', alpha=0.7)
plt.xlabel("X qiymatlar oralig'i")
plt.ylabel("n_i / h")
plt.title("Chastotalar Gistogrammasi")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
Nisbiy chastotalar poligoni
```

```python
x_i = [2, 4, 5, 7, 10]
w_i = [0.15, 0.2, 0.1, 0.1, 0.45]
# Ma'lumotlar jadvalini tashkil etish
df = pd.DataFrame({
    'x_i' : x_i,
    'w_i': w_i
})
df.index = list(range(1, len(x_i) + 1))
df.plot( x='x_i', y='w_i',color='r' figsize=(14,6))
plt.title('Nisbiy chastotalar poligoni')
plt.xlabel('X o\'qi')
plt.ylabel('Nisbiy chastotalar')
df
```

Nisbiy chastotalar gistogrammasi

```python
bosh_qiymat = 2
qadam = 5
sikl = 5
arr = []
for i in range(sikl):
    arr.append(bosh_qiymat+qadam)
    bosh_qiymat += qadam
n_i= [5, 10 ,25, 6, 4]
n_i_dev_h = [i/qadam for i in n_i]

df = pd.DataFrame({
    'x_i-x_(i+1)' : arr,
    'n_i' : n_i,
    'n_i/h' : n_i_dev_h    
})
df.index=np.arange(1, len(n_i) + 1)
plt.bar(df['x_i-x_(i+1)'], df['n_i/h'], width=qadam, align='edge', edgecolor='red', alpha=0.7, label="Chastotalar ustunlari")
plt.xlabel('X o\'qi')
plt.ylabel('Nisbiy chastotalar')
plt.title('Nisbiy chastotalar gistogrammasi')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.legend(loc='upper right')
df
```
